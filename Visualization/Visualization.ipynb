```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Crimerate.csv')  ##Indlæser csv datasættet
```

```python
data  
```

```python
data2 = data[['Category', 'Descript','DayOfWeek','Date','Time','PdDistrict','Resolution']].copy()
#Her udvælger vi de relevante variabler
```

```python
data2
```

```python
numbers_pr_categori = data2["Category"].value_counts() #Tæller antallet af crimes i hver kategori
numbers_pr_categori = pd.DataFrame(numbers_pr_categori) #sætter i en tabel
```

```python
numbers_pr_categori = numbers_pr_categori.reset_index() #Her giver vi kategorierne et index
```

```python
numbers_pr_categori.rename(columns={'index': 'Type_of_crime'}, inplace=True) #Omdøber index til "Type_of_crime"
numbers_pr_categori.rename(columns={'Category': 'Amount'}, inplace=True) #Omdøber category til "Amount"
```

```python
numbers_pr_categori
```

```python
numbers_pr_categori.plot(x='Type_of_crime', y='Amount', kind='bar', figsize=(16,10)) 
#Her dannes der et histogram, hvor x = type af kriminalitet og y = antal der er begået
```

```python
numbers_pr_DayOfWeek = data2["DayOfWeek"].value_counts()  #De næste kommandoer gøres det samme med day/antal
numbers_pr_DayOfWeek = pd.DataFrame(numbers_pr_DayOfWeek)
numbers_pr_DayOfWeek = numbers_pr_DayOfWeek.reset_index()
```

```python
numbers_pr_DayOfWeek.rename(columns={'index': 'Day_of_week'},inplace=True)
numbers_pr_DayOfWeek.rename(columns={'DayOfWeek': 'Amount'},inplace=True)
```

```python
numbers_pr_DayOfWeek
```

```python
numbers_pr_DayOfWeek.plot(x='Day_of_week',y='Amount',kind='bar',figsize=(16,10)) 
#Histrogram der viser antal crimes der er begået på de forskellige ugedage
```

```python
counts_pr_category = data2.groupby(by=["Category", "DayOfWeek"]).size().reset_index(name="counts")
counts_pr_category
# Disse tabeller viser fordellingen af antallet af criminalitet typer på de forskellige ugedage.
```

```python
green_diamond = dict(markerfacecolor='g', marker='D')  #Change outliers to diamond symbol
plt.boxplot(numbers_pr_categori.Amount,flierprops=green_diamond)
plt.title("BoxPlot for crimes begået af hver kategori")
# Boxplottet over antal crimes fordelt på catagories 
```

```python
plt.boxplot(numbers_pr_DayOfWeek.Amount)
plt.title("BoxPlot for crimes begået på hverdagen")
#boxplot der viser criminalitet begået på de forskellige ugedage
```

```python
#Fordelling af antal crimes begået i hver categori vist i % 
labels = numbers_pr_categori["Type_of_crime"]
plt.pie(numbers_pr_categori.Amount, labels = labels,radius = 3, shadow =True,autopct='%.1f%%' )
plt.title('Kriminalitet begået pr. kategori', y=2, fontsize=30)
plt.legend(bbox_to_anchor =(3,-1), loc='lower center')
```

```python
#Fordelling af antal crimes begået på hvilket hverdage vist i % 
labels = numbers_pr_DayOfWeek["Day_of_week"]
plt.pie(numbers_pr_DayOfWeek.Amount,labels = labels,autopct='%.1f%%',radius =2)
plt.legend(bbox_to_anchor =(2,0), loc='lower center')
plt.title("Kriminalitet fordelt på ugedage",fontsize=30,y=1.5)
```

```python
a,b=np.polyfit(numbers_pr_categori.index,numbers_pr_categori.Amount,1)
plt.scatter(numbers_pr_categori.index,numbers_pr_categori.Amount)
plt.plot(numbers_pr_categori.index,a*numbers_pr_categori.index+b)
#Scatterplot af antal crimes / kategorier, hvor man kan se at der ikke
#er en linear sammenhæng
```

```python
a,b=np.polyfit(numbers_pr_DayOfWeek.index,numbers_pr_DayOfWeek.Amount,1)
plt.scatter(numbers_pr_DayOfWeek.index,numbers_pr_DayOfWeek.Amount)
plt.plot(numbers_pr_DayOfWeek.index,a*numbers_pr_DayOfWeek.index+b)
#Scatterplot over antal crimes / ugedage, hvor man kan se en linear sammenhæng
```
